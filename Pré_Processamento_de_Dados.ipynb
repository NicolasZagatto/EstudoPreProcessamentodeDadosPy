```python
import pandas as pd
```

```python
dic = {'Nome': ['Isa', 'Nicolas', 'Carlos'],
       'Idade':['19', '28', '18'],
       'Salário':['19000', '5000', '2500'],
       'Gênero':['F', 'M', 'M'],
       'Cargo':['Gerente', 'Supervisor', 'Estagiário']
}
```

```python
df = pd.DataFrame(dic)
df
```

```python
df = pd.get_dummies(df, columns=['Gênero'])
```
